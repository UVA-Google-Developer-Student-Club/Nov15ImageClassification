```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./Shoe.JPG')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
print(img2.shape)
```

```python
center = (img2.shape[0] // 2, img2.shape[1] // 2)
sideLength = 2100
print(center)
croppedImg = img2[center[0] - sideLength//2 : center[0] + sideLength//2, center[1] - sideLength//2 : center[1] + sideLength//2,:]
plt.imshow(croppedImg)
```

```python
# hsv = cv2.cvtColor(croppedImg, cv2.COLOR_RGB2HSV)
# plt.imshow(hsv)
mask = cv2.inRange(croppedImg, numpy.array([100, 100, 90]), numpy.array([255, 255, 255]))
mask = 255 - cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)
plt.imshow(mask)

```

```python
final = cv2.resize(mask, (28, 28), interpolation = cv2.INTER_AREA)
plt.figure(figsize=(1,1))
plt.imshow(final)
# save to files
cv2.imwrite('final_shoe.png', cv2.cvtColor(final, cv2.COLOR_RGB2GRAY))
```
